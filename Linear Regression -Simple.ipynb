```
# Build a model for the Salary vs Experience dataset

# x = Feature = Experience (Independent)
# y = Target = Salary (Dependent)

import numpy as np
np.random.seed(3)
x = np.arange(1,9)
x = x.reshape(-1, 1)
y = np.random.randint(45000,200000,8)
y = np.sort(y)
x
```

```
x.ndim
```

```
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

# Learn how to solve this error
# **ValueError: Expected 2D array, got 1D array instead:**

```
# Sklearn - ML library in Python
# ML ALgorithm - Family Library

#Syntax: from sklearn.family_library import Algorithm_name/Module

from sklearn.linear_model import LinearRegression
```

```
model = LinearRegression()
model.fit(x,y)  # Maps input values with output values for getting the pattern
```

```
# y = mx+c
# Slope(coef)
# yintercept
m = model.coef_
m
```

```
c = model.intercept_
c
```

```
# Best Fit Line
y_pred = model.predict(x)
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.plot(x,y_pred,c='r')
plt.show()
```
